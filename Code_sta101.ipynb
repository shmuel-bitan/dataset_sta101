# **Identifiez les clusters dans les données (K-means algorithm)**

L'algorithme K-Means, une technique puissante d'apprentissage automatique non supervisé, sert d'outil polyvalent pour le regroupement de données et la reconnaissance de motifs. En fonctionnant sur le principe de la partition d'un ensemble de données en groupes distincts basés sur la similarité, K-Means s'efforce de minimiser les distances entre les points d'un même cluster.

Pour notre ensemble de données Goodreads, nous utilisons l'algorithme pour catégoriser les livres partageant des caractéristiques similaires dans le même cluster, ce qui permet d'explorer les tendances et les motifs sous-jacents au sein du paysage littéraire. En mettant en œuvre K-Means sur cet ensemble de données spécifique, nous visons à découvrir les structures et les relations inhérentes entre les livres.

```r
# Charger les bibliothèques nécessaires
library(dplyr)
library(tidyr)
library(scales)

# Définir les caractéristiques numériques et les caractéristiques encodées en one-hot
numerical_features <- c('rating', 'pages', 'numRatings', 'likedPercent', 'bbeScore', 'bbeVotes', 'price')
one_hot_encoded_features <- setdiff(names(goodreads_data), numerical_features)

# Normaliser uniquement les caractéristiques numériques
goodreads_data_scaled_numerical <- goodreads_data %>%
  select(all_of(numerical_features)) %>%
  mutate(across(everything(), ~ scale(.)[, 1]))  # Utilise scale() pour normaliser

# Combiner les caractéristiques numériques normalisées avec les caractéristiques encodées en one-hot
goodreads_data_scaled <- cbind(goodreads_data_scaled_numerical, goodreads_data %>% select(all_of(one_hot_encoded_features)))

# Afficher les premières lignes du jeu de données normalisé
head(goodreads_data_scaled)
```

Le choix du nombre de clusters
Pour déterminer combien de clusters utiliser, une méthode courante est la méthode du "coude" (Elbow method). Cette stratégie nous aide à déterminer le nombre optimal de clusters en analysant la relation entre le nombre de clusters (K) et la somme des carrés intra-cluster (WCSS).

Nous exécutons K-Means pour différentes valeurs de K, puis traçons le WCSS résultant. Un point "coude", ou changement dans le profil, est identifié à partir duquel l'ajout de clusters supplémentaires offre des rendements décroissants dans la réduction du WCSS. Ce nombre de clusters représente un équilibre entre la capture de motifs significatifs et l'évitement d'une granularité excessive, guidant la sélection d'une valeur K appropriée pour un clustering efficace.

```r
# Vérifiez les valeurs NA, NaN, et Inf dans les données
summary(goodreads_data_scaled_numerical)

# Remplacer les valeurs NA et NaN par la moyenne des colonnes
goodreads_data_scaled_numerical <- goodreads_data_scaled_numerical %>%
  mutate_all(~ifelse(is.na(.) | is.nan(.), mean(., na.rm = TRUE), .))

# Remplacer les valeurs Inf par la valeur maximale des colonnes
goodreads_data_scaled_numerical <- goodreads_data_scaled_numerical %>%
  mutate_all(~ifelse(is.infinite(.), max(., na.rm = TRUE), .))
```

```r
# Initialiser un vecteur pour stocker les valeurs de WCSS (Within-Cluster Sum of Squares)
wcss <- numeric()

# Définir la plage de valeurs de K à tester
testing_range <- seq(1, 15, by = 2)

# Calculer WCSS pour chaque valeur de K dans testing_range
for (k in testing_range) {
  kmeans_model <- kmeans(goodreads_data_scaled_numerical, centers = k, nstart = 10)
  wcss <- c(wcss, kmeans_model$tot.withinss)
}

# Tracer le graphique du coude pour déterminer le nombre optimal de clusters
plot(testing_range, wcss, type = "b", pch = 19, col = "blue",
     xlab = "Nombre de clusters", ylab = "WCSS",
     main = "Méthode du coude pour le choix du nombre optimal de clusters")
```

Au vue de la méthode du coude, on initialise le nombre de clusters à 3

```r
library(dplyr)
library(tidyr)
library(scatterplot3d)

# Sélectionner et scaler les données numériques
goodreads_data_scaled_numerical <- goodreads_data %>%
  select(pages, rating, numRatings) %>%
  mutate_all(~ifelse(is.na(.) | is.nan(.), mean(., na.rm = TRUE), .)) %>%
  mutate_all(~ifelse(is.infinite(.), max(., na.rm = TRUE), .)) %>%
  scale()

# Vérifier s'il y a encore des NA, NaN ou Inf
summary(goodreads_data_scaled_numerical)

# Définir le nombre de clusters
k <- 3

# Exécuter K-Means
kmeans_result <- kmeans(goodreads_data_scaled_numerical, centers = k, nstart = 10)

# Récupérer les étiquettes de cluster
cluster_labels <- kmeans_result$cluster

# Créer un graphique 3D des clusters
colors <- rainbow(k)  # Définir une palette de couleurs pour les clusters

# Utiliser scatterplot3d pour créer le graphique
plot <- scatterplot3d(goodreads_data_scaled_numerical[, "pages"],
                      goodreads_data_scaled_numerical[, "rating"],
                      goodreads_data_scaled_numerical[, "numRatings"],
                      color = colors[cluster_labels],
                      pch = 16,
                      main = "Clusters de données",
                      xlab = "Number of Pages",
                      ylab = "Average Rating",
                      zlab = "Number of Ratings")

# Ajouter une légende
legend("topright", legend = paste("Cluster", 1:k), col = colors, pch = 16, cex = 1.2)

# Afficher le graphique
plot
```

```r
library(dplyr)
library(tidyr)

# Ajouter les étiquettes de cluster au dataframe original
goodreads_data <- goodreads_data %>%
  mutate(cluster_label = cluster_labels)

# Calculer les statistiques descriptives pour chaque cluster
cluster_summary <- goodreads_data %>%
  group_by(cluster_label) %>%
  summarise(
    count = n(),
    avg_rating = mean(rating, na.rm = TRUE),
    avg_pages = mean(pages, na.rm = TRUE),
    avg_numRatings = mean(numRatings, na.rm = TRUE),
    avg_likedPercent = mean(likedPercent, na.rm = TRUE),
    avg_bbeScore = mean(bbeScore, na.rm = TRUE),
    avg_bbeVotes = mean(bbeVotes, na.rm = TRUE),
    avg_price = mean(price, na.rm = TRUE)
  )

# Afficher le résumé des clusters
print(cluster_summary)

# Calculer les genres les plus répandus dans chaque cluster
genre_summary <- goodreads_data %>%
  group_by(cluster_label) %>%
  unnest(genre = strsplit(as.character(genres), ", ")) %>%  # Séparer les genres multiples
  count(cluster_label, genre) %>%
  group_by(cluster_label) %>%
  slice_max(n, n = 5)  # Sélectionner les 5 genres les plus répandus

# Afficher les genres les plus répandus pour chaque cluster
print(genre_summary)

# Calculer les langues les plus répandues dans chaque cluster
language_summary <- goodreads_data %>%
  group_by(cluster_label) %>%
  count(cluster_label, language) %>%
  group_by(cluster_label) %>%
  slice_max(n, n = 5)  # Sélectionner les 5 langues les plus répandues

# Afficher les langues les plus répandues pour chaque cluster
print(language_summary)
```

```r
# Charger les bibliothèques nécessaires
library(FactoMineR)
library(factoextra)
library(dplyr)

# Prendre un échantillon aléatoire de 1000 lignes du jeu de données
set.seed(123)  # Pour la reproductibilité
sample_data <- goodreads_data %>% sample_n(1000)

# Sélectionner uniquement les variables qualitatives pour l'ACM
qualitative_vars <- sample_data %>%
  select(language, genres, author) %>%
  na.omit()  # Supprimer les lignes contenant des valeurs manquantes

# Vérifier les dimensions après suppression des valeurs manquantes
dim(qualitative_vars)

# Réaliser l'Analyse des Correspondances Multiples (ACM)
acm_result <- MCA(qualitative_vars, graph = FALSE)

# Afficher le résumé des résultats de l'ACM
summary(acm_result)

# Visualiser les résultats de l'ACM
fviz_mca_ind(acm_result,
             label = "none", # Masquer les étiquettes des individus
             habillage = "genres", # Colorer par genres
             addEllipses = TRUE, # Ajouter des ellipses de confiance
             ellipse.type = "confidence"
             )

fviz_mca_var(acm_result,
             repel = TRUE # Éviter le chevauchement du texte
             )

# Contribution des variables
fviz_contrib(acm_result, choice = "var", axes = 1:2)

# Valeurs propres / Variances
eig_values <- get_eigenvalue(acm_result)
fviz_screeplot(acm_result, addlabels = TRUE, ylim = c(0, 50))
```
